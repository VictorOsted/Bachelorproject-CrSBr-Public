```python
def plotter(unpol, pol_yy, pol_zz, eres, d1, plot_dispersion=False, plot_cut=False, plot_gap=False, plot_eres_d1=True):
    # Only plots the dispersion if plot_dispersion is True
    if plot_dispersion:
        def colorplot(Y, Z, vmin, vmax, label, eres, d1):
            X = np.linspace(0.5, 2.5, 500)
            plt.pcolormesh(X, Y, Z, shading='gouraud', vmin=vmin, vmax=vmax)
            plt.colorbar()
            plt.ylabel('Energy (meV)')
            plt.xlabel('(0.75,k,0) [h,k,l]')
            #plt.title(f'LSWT on CrSBr, Eres={eres} meV, D1={d1} meV')
            plt.text(0.02, 0.98, label, transform=plt.gca().transAxes, 
                    verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        plt.figure()
        colorplot(energies, unpol, vmin=0, vmax=1, label='Unpolarized', eres=eres, d1=d1)
        #plt.savefig(f'unpol, Eres={eres}, D1={d1}.png')
        plt.figure()
        colorplot(energies, pol_yy, vmin=0, vmax=1, label='Polarized (M_yy)', eres=eres, d1=d1)
        #plt.savefig(f'pol_yy, Eres={eres}, D1={d1}.png')
        plt.figure()
        colorplot(energies, pol_zz, vmin=0, vmax=1, label='Polarized (M_zz)', eres=eres, d1=d1)
        #plt.savefig(f'pol_zz, Eres={eres}, D1={d1}.png')

    # Defines the cuts and the x_mask for plotting
    cut_unpol = unpol[:, 240:261].mean(axis=1)
    cut_yy = pol_yy[:, 240:261].mean(axis=1)
    cut_zz = pol_zz[:, 240:261].mean(axis=1)
    x_mask = (energies > 30) & (energies < 45)

    # Only plots the cuts if plot_cut is True
    if plot_cut:
        plt.figure()
        plt.plot(energies[x_mask], cut_unpol[x_mask], label='Unpolarized')
        plt.plot(energies[x_mask], cut_yy[x_mask], label='Polarized (M_yy)')
        plt.plot(energies[x_mask], cut_zz[x_mask], label='Polarized (M_zz)')
        plt.xlabel('Energy (meV)')
        plt.ylabel('Intensity (a.u.)')
        #plt.title(f'Eres={eres} meV, D1={d1} meV, (0.75,k,0)')
        plt.grid(alpha=0.3)
        plt.legend()

    peak_indices, _ = find_peaks(cut_unpol, height=0.1) # Finds the peaks in the unpolarized cut (same for pol_yy and pol_zz)
    gap = None
    # If there is a gap it calculates the gap energy and plots lines at the peak if plot_cut is True
    if len(peak_indices) > 1:
        gap = energies[peak_indices[1]] - energies[peak_indices[0]]
        if plot_gap:
            plt.axvline(x=energies[peak_indices[0]], color='red', linestyle='--', alpha=0.5,
                        label=f'Energy gap: {gap:.2f} meV')
            plt.axvline(x=energies[peak_indices[1]], color='red', linestyle='--', alpha=0.5)
    # If there is only one peak it plots a single line at the peak if plot_cut is True
    elif len(peak_indices) == 1 and plot_gap:
        plt.axvline(x=energies[peak_indices[0]], color='red', linestyle='--', alpha=0.5,
                    label='Single peak (no gap)')
    if plot_gap:
        plt.legend()

    if plot_eres_d1 and gap is not None:
        ax = plt.gca()
        gap_label = 'Gap' if not any(line.get_label() == 'Gap' for line in ax.lines) else '_nolegend_'
        plt.plot(d1, gap, marker='o', linestyle='', color='red', label=gap_label)
        plt.xlabel('D$_1$ (meV)')
        plt.ylabel('Energy gap (meV)')
        plt.xlim(0, 0.75)
        plt.ylim(0, 5)
        #plt.title(f'Energy gap vs D1 with Eres={eres} meV')
        plt.grid(alpha=0.3)
    elif plot_eres_d1 and gap is None:
        plt.plot(d1, np.nan, marker='o', linestyle='', color='red')
        plt.xlabel('D$_1$ (meV)')
        plt.ylabel('Energy gap (meV)')
        plt.xlim(0, 0.75)
        plt.ylim(0, 5)
        #plt.title(f'Energy gap vs D1 with Eres={eres} meV')
        plt.grid(alpha=0.3)

    return cut_unpol, x_mask, gap
```

```python
def gaussian(cut_unpol, cut_yy, cut_zz, eres, d1):
    def double(E, mu, std, mu2, std2,c,d):
        return (c/np.sqrt(2*np.pi*std**2))*(np.exp(-((E-mu)**2)/(2*std**2))) + (d/(2*np.pi*std2**2))*(np.exp(-((E-mu2)**2)/(2*std2**2)))
    

    unpol, x_mask, gap = plotter(cut_unpol, cut_yy, cut_zz, eres=eres, d1=d1)
    error = np.ones(len(energies[x_mask]))*0.01
    pard, covd = curve_fit(double, energies[x_mask], unpol[x_mask], p0 = [36, 0.2, 38, 0.2, 1,1], absolute_sigma= True, sigma = error, maxfev = 100000)

    errord = np.sqrt(np.diag(covd))

    return errord, gap


def linear_fit(gaps, errord, d1_values, plot_fit=True):
    def lin(x, a, b):
        return a * x + b

    prop_error = np.asarray([np.sqrt(e[0]**2 + e[2]**2) if (e is not None) else np.nan for e in errord], dtype=float)
    gaps = np.asarray(gaps, dtype=float)

    X = np.asarray(d1_values, dtype=float)

    valid = np.isfinite(X) & np.isfinite(gaps) & np.isfinite(prop_error) & (gaps != 0) & (prop_error > 0)
    X_fit = X[valid]
    gaps_fit = gaps[valid]
    prop_error_fit = prop_error[valid]

    par, cov = curve_fit(lin, X_fit, gaps_fit, p0=[1, 0], absolute_sigma=True, sigma=prop_error_fit, maxfev=100000)
    Y = lin(X_fit, *par)

    y_diff = gaps_fit - Y

    chi = np.sum((y_diff)**2/prop_error_fit**2)

    red_chi = chi/(len(X_fit) - len(par))

    print(f"Reduced Chi-squared: {red_chi:.2f}")

    if plot_fit:
        plt.errorbar(X_fit, gaps_fit, yerr= prop_error_fit, fmt = 'none', label = 'error', zorder = 1)
        plt.plot(X_fit, Y, color='orange', label='Linear fit')
        plt.legend()
```
