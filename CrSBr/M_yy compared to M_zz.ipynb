```python
def channels(unpol, pol_yy, pol_zz, Eres, D1):
    def colorplot(Y, Z, vmin, vmax, label, Eres, D1):
        X = np.linspace(0.5, 2.5, 500)
        plt.pcolormesh(X, Y, Z, shading='gouraud', vmin=vmin, vmax=vmax)#, cmap='bwr')
        plt.colorbar()
        plt.ylabel('Energy (meV)')
        plt.xlabel('(0.75,k,0) [h,k,l]')
        plt.title(f'LWST on CrSBr, Eres={Eres} meV, D1={D1}')
        plt.text(0.02, 0.98, label, transform=plt.gca().transAxes, 
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
    def intensity_average(x, y, unpol, pol_yy, pol_zz, Eres, D1):
        plt.figure()
        int_cut_range = unpol[:, 60:186].mean(axis=1)
        plt.plot(energies, int_cut_range,'.', label='Unpolarized')
        int_cut_range_yy = pol_yy[:, 60:186].mean(axis=1)
        plt.plot(energies, int_cut_range_yy,'.', label='Polarized (M_yy)')
        int_cut_range_zz = pol_zz[:, 60:186].mean(axis=1)
        plt.plot(energies, int_cut_range_zz,'.', label='Polarized (M_zz)')
        plt.xlabel('Energy (meV)')
        plt.ylabel('Intensity (a.u.)')
        plt.title(f'Eres={Eres} meV, D1={D1}, (0.75,k,0)')
        plt.grid(alpha=0.3)
        plt.legend()
        return int_cut_range_yy, int_cut_range_zz
    
    # plt.figure()
    # colorplot(energies, unpol, vmin=0, vmax=1, label='Unpolarized', Eres=Eres, D1=D1)
    # plt.figure()
    # colorplot(energies, pol_yy, vmin=0, vmax=1, label='Polarized (M_yy)', Eres=Eres, D1=D1  )
    # plt.figure()
    # colorplot(energies, pol_zz, vmin=0, vmax=1, label='Polarized (M_zz)', Eres=Eres, D1=D1)
    
    b, c = intensity_average(energies, unpol, unpol, pol_yy, pol_zz, Eres=Eres, D1=D1)
    return b, c
```
