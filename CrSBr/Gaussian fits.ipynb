```python
def gauss(E, mu, std,c):
    return (c/np.sqrt(2*np.pi*std**2))*(np.exp(-((E-mu)**2)/(2*std**2)))

def double(E, mu, std, mu2, std2,c,d):
    return (c/np.sqrt(2*np.pi*std**2))*(np.exp(-((E-mu)**2)/(2*std**2))) + (d/(2*np.pi*std2**2))*(np.exp(-((E-mu2)**2)/(2*std2**2)))
```

```python
def gaussian(cut_unpol, cut_yy, cut_zz, E_res, D1):
    x_mask= (energies > 30) & (energies < 45)
    data = channels(cut_unpol, cut_yy, cut_zz, Eres=E_res, D1=D1)
    error = np.ones(len(energies[x_mask]))*0.001
    pard, covd = curve_fit(double, energies[x_mask], data[x_mask], p0 = [36, 0.2, 38, 0.2, 1,1], absolute_sigma= True, sigma = error, maxfev = 100000)
    parg, covg = curve_fit(gauss, energies[x_mask], data[x_mask], p0 = [37, 0.8,2], absolute_sigma= True, sigma= error, maxfev = 100000)

    X = np.linspace(np.min(energies[x_mask]), np.max(energies[x_mask]), len(energies[x_mask]))
    YD = double(X, *pard)
    YG = gauss(X, *parg)

    plt.errorbar(energies[x_mask], data[x_mask], yerr= error, fmt = 'none', label = 'error', zorder = 0)
    plt.plot(X,YD, color = 'purple', label = 'double gaussian')
    plt.plot(X, YG, label = 'single gaussian')
    plt.legend()

    y_diff_d = data[x_mask] - YD
    y_diff_g = data[x_mask] - YG

    chi_d = np.sum((y_diff_d)**2/error**2)
    chi_g = np.sum((y_diff_g)**2/error**2)

    red_chi_d = chi_d/(len(energies[x_mask]) - len(pard))
    red_chi_g = chi_g/(len(energies[x_mask]) - len(parg))

    errord = np.sqrt(np.diag(covd))
    errorg = np.sqrt(np.diag(covg)) 
    
    return red_chi_d, red_chi_g


    
```
