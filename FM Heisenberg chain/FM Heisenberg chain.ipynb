```python
# Imports necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Loads the energies and intensities from the .txt files generated by Sunny
energies = np.genfromtxt(os.path.join(os.path.dirname(os.getcwd()), 'FM Heisenberg chain/FM Heisenberg chain energies.txt'), delimiter=',')
data = np.genfromtxt(os.path.join(os.path.dirname(os.getcwd()), 'FM Heisenberg chain/FM Heisenberg chain intensities.txt'), delimiter=',')
```

```python
# Makes a colorplot of the intensities over a given path in reciprocal space
def colorplot(Y,Z):
    X = np.linspace(0,1,500)
    plt.pcolormesh(X,Y,Z,shading='gouraud')
    plt.colorbar()
    plt.ylabel('Energy (meV)')

# Calculates the theoretical spin wave dispersion for a ferromagnetic Heisenberg chain
def theory(q,J,S,a):
    return 2*J*S*(1-np.cos(q*a))
```

```python
# Plots the simulated intensities and the theoretical dispersion
colorplot(energies, data)
a = 3
x = np.linspace(0,1,500)
x_rlu = np.linspace(0,2*np.pi/a,500)
plt.plot(x,theory(x_rlu,1,3/2,a), color='red', label='Theory', ls='--')
plt.xlabel('(h,0,0) [h,k,l]')
plt.legend()
plt.title('Simulated and theoretical intensity of the FM Heisenberg chain')
```
